```python
X = df[['Versus','Ground']].values
Y = df['S/R'].values
#print(len(Y))
#lab_enc = preprocessing.LabelEncoder()
#Y1= lab_enc.fit_transform(Y)
Y=list(Y)
#print(Y)
for i in range(len(Y)):
    if Y[i]>83and Y[i]<=124:
        Y[i]='select'
    elif Y[i]>42 and Y[i]<=83:
        Y[i]='improvement_required'
    elif Y[i]>=0 and Y[i]<=42:
        Y[i]='Not_sel'
    
print(Y)
#lab_enc = preprocessing.LabelEncoder()
#Y1= lab_enc.fit_transform(Y)
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(X, Y, train_size=0.8, random_state=1)

```

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_inputs,train_classes)
dtc.score(test_inputs, test_classes)
#print(dtc.predict([[7,39]]))
#print(dtc.predict([[7,39]]))
```

```python
dot_data = StringIO()
export_graphviz(dtc, out_file=dot_data, filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
